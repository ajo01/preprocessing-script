```python
dep_col = df.iloc[:,1]
```

```python
std = dep_col.std()
std
```

```python
mean = dep_col.mean()
mean
```

```python
max_dep_allowed = 4 * std + mean
max_dep_allowed
```

```python
max_dep_allowed = math.ceil(max_dep_allowed)
max_dep_allowed
```

```python
filtered_df = df[df['Dem_isolation_kids'] <= max_dep_allowed]
```
